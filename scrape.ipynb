```python
from datetime import datetime, timedelta
import time

x_token = "dd279566d25e9b2a33f82e1bb43c0effac57d80b"
```

# Primer intento 2024

```python
filename_prefix = 'sh_'
# Date ranges and search keyword template
start_date = datetime(2024, 10, 1)  # Start date
end_date = datetime(2024, 12, 31)  # End date
delta = timedelta(days=5)  # 5-day interval

# Loop through date ranges
current_date = start_date
while current_date < end_date:
    # Define date range for this iteration
    next_date = current_date + delta
    formatted_start = current_date.strftime("%Y-%m-%d")
    formatted_end = next_date.strftime("%Y-%m-%d")
    
    # Ensure the last date doesn't exceed the specified end date
    if next_date > end_date:
        formatted_end = end_date.strftime("%Y-%m-%d")
    
    # Generate filename and search keyword
    filename = f"{filename_prefix}{formatted_start}_to_{formatted_end}.csv"
    search_keyword = f"sheinbaum until:{formatted_end} since:{formatted_start}"
    limit = 1000  # Adjusted limit

    # Run tweet-harvest command
    print(f"Fetching data from {formatted_start} to {formatted_end}...")
    !npx --yes tweet-harvest@latest -o "{filename}" -s "{search_keyword}" -l {limit} --token {x_token}
    
    # Pause for 15 minutes
    print("Pausing for 15 minutes...")
    time.sleep(910)
    
    # Move to the next date range
    current_date = next_date

print("Data fetching complete.")
```
